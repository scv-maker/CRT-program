# print all natural even numbers upto n


```
n=int(input("enter the value"))
i=1
while i<=n:
  if (i%2) == 0:
    print(i,end="  ")
  i= i+1
```

# addition of even numbers upto n

```
n=int(input("enter the number"))
i=1
sum = 0
while i<=n:
  if i%2 == 0 :
   sum = sum+i
  i= i + 1
print(sum)
```
